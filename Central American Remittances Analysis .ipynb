```python
df = pd.read_excel('rpw_dataset_2011_2022_q1_0.xlsx', sheet_name = 5)
```

```python
caCountries = ["Guatemala","Honduras","Nicaragua","El Salvador","Costa Rica","Panama","Belize"]
```

```python
dfCA = df.loc[df['destination_name'].isin(caCountries)]
```

```python
dfCA = dfCA[['period', 'source_code', 'destination_name', 'pickup method', 'pickup location', 
             'firm', 'firm_type','speed actual', 'cc1 total cost %', 
             'transparent', 'receiving network coverage']]
```

```python
print("Descriptive stats on % of total transactaction cost in fees for Remittances from the US to CA")
print("Transfer amount = 200")

dfCA['cc1 total cost %'].describe()
```

```python
def getCountryData(countryName):
    df = dfCA.loc[dfCA['destination_name'] == countryName]
    return df

def printCountryData(countries):
    print("Descriptive stats on % of total transactaction cost in fees for remittances from the US")
    print("Transfer amount = $200")
    for country in countries:
        resultData = getCountryData(country)
        print("Data for ", country, ": \n")
        print(resultData['cc1 total cost %'].describe(), "\n")
        
```

```python
printCountryData(caCountries)
```
